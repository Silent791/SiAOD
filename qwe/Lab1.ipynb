### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер

```python
import random
import time
from random import randint
```

```python
import random
import time
from random import randint
user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit
print()


def mgenerate(m, n, min_l, max_l):
    mat = [] 
    for i in range (int(m)): 
        row=[]
        for j in range (int(n)):
            randNumber = random.randint(int(min_l), int(max_l))
            row.append (randNumber)
        mat.append(row)
    return mat
def mprint(mat):    
    for i in range(0, len(mat)):
        for j in range(0, len(mat[i])):
            print(mat[i][j], end=' ')
        print()
mat=mgenerate(user_m, user_n, user_min_limit, user_max_limit)
mprint(mat)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
M = copy.deepcopy(mat)
def choice_sort(row):
    for i in range(len(row) - 1):
        m = i
        j = i + 1
        while j < len(row):
            if row[j] < row[m]:
                m = j
            j = j + 1
        row[i], row[m] = row[m], row[i]
for row in M:
    row = choice_sort(row)
mprint(M)
choice_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(choice_sort_time))
```

```python
# Сортировка вставкой.
start_time = time.time()
M = copy.deepcopy(mat)
def insert_sort(row):
    for i in range(len(row)):
        cursor = row[i]
        pos = i
        while pos > 0 and row[pos - 1] > cursor:
            row[pos] = row[pos - 1]
            pos = pos - 1
        row[pos] = cursor
    return row
for row in M:
    row = insert_sort(row)
mprint(M)
insert_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(insert_sort_time))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
M = copy.deepcopy(mat)
def exch_sort(row):
    for i in range (len(row)-1):
        for j in range(len(row)-i-1):
            if row[j] > row[j+1]:
                row[j], row[j+1] = row[j+1], row[j]
    return row
for row in M:
    row = exch_sort(row)
mprint(M)
exch_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(exch_sort_time))
```

```python
# Сортировка Шелла.
start_time = time.time()
M = copy.deepcopy(mat)
def sh_sort(row):
    step = len(row) // 2
    while step > 0:
        for i in range(step, len(row)):
            position = i
            value = row[i]
            while position >= step and row[position - step] > value:
                row[position] = row[position - step]
                position -= step
                row[position] = value
        step = step // 2
for row in M:
    row = sh_sort(row)
mprint(M)
sh_sort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(sh_sort_time))
```

```python
# Быстрая сортировка.
start_time = time.time()
def qsort(row):
    if len(row)<=1:
        return row
    q=row[0]
    l=[]
    c=[]
    r=[]
    for i in range (len(row)):
        if row[i]<q:
            l.append(row[i])
        if row[i]==q:
            c.append(row[i])
        if row[i]>q:
            r.append(row[i])
    return qsort(l) + c + qsort(r)
M = copy.deepcopy(mat)
M1 = []
for row in M:
    M1.append(qsort(row))
mprint(M1)
qsort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(qsort_time))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
import heapq
M = copy.deepcopy(mat)
def tsort(row, hsize):
    heap = []
    if len(row) >= hsize:
        heap = row[:hsize]
    else:
        heap = row[:]
    heapq.heapify(heap)
    
    win = [float('-inf')]
    los = []
    i = hsize
    while i < len(row) or len(heap) > 0:
        top = heapq.heappop(heap)
        if top >= win[-1]:
            win.append(top)
        else:
            los.append(top)
            
        if i < len(row):
            heapq.heappush(heap, row[i])
        i += 1
        
    del win[0]
    row.clear()
    row.extend(los)
    row.extend(win)
    
    if len(los) > 0:
        tsort(row, hsize)
        
for i in range(len(M)):
        tsort(M[i], 10)
mprint(M)

tsort_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(tsort_time))

### Вывод
```
